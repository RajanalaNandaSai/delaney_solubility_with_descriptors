#**My first ML project**

```python
import pandas as pd
df= pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv")
df
```

## **Data preparation**

### **data separation as x and y**

```python
y= df['logS']
y
```

--> to separate x we use "drop" command to eliminate y from the data set && axis=0 work in row mode and axis =1 work in column mode

```python
x= df.drop('logS', axis=1)
x
```

### **data splitting into train & test sets**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

##**Model building**

###**linear regression**

#### **training model**

```python
from sklearn.linear_model import LinearRegression

lr= LinearRegression()
lr.fit(x_train,y_train)
```

####**Prediction by the model**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

####**model evaluation**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('lr_train_mse:',lr_train_mse)
print('lr_train_r2:',lr_train_r2)
print('lr_test_mse:',lr_test_mse)
print('lr_test_r2:',lr_test_r2)
```

```python
lr_results = pd.DataFrame(['linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns= ['method','training mse','training r2','test mse', 'test r2']
lr_results
```

###**Random forest regression**

####**training model**

```python
from sklearn.ensemble import RandomForestRegressor

rf= RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train,y_train)
```

####**prediction**

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

####**model evaluation**

```python
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
print('rf_train_mse:',rf_train_mse)
print('rf_train_r2:',rf_train_r2)
print('rf_test_mse:',rf_test_mse)
print('rf_test_r2:',rf_test_r2)
```

```python
rf_results = pd.DataFrame(['random forest regression', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns= ['method','training mse','training r2','test mse', 'test r2']
rf_results
```

###**model comparision**

```python
df_models= pd.concat([lr_results,rf_results],axis=0).reset_index(drop=True) #.reset_index(drop=True) this is written to avid 0 sno in both rows assigning proper index

df_models
```
